```python
def nan_impute(main_data,variable,median):
        main_data[variable +'_random']=main_data[variable]
        random_sample=main_data[variable].dropna().sample(main_data[variable].isnull().sum(),random_state=0)
        random_sample.index=main_data[main_data[variable].isnull()].index
        main_data.loc[main_data[variable].isnull(),variable +'_random']=random_sample
        main_data.drop([variable],inplace=True,axis=1)
```

# KNN for finding the categorical missing data

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
def KNN (x_train,x_test,y_train,dataset,feature):
        scaler=StandardScaler()
        scaler.fit(x_train)
        x_train_scaled=scaler.transform(x_train)
        x_test_scaled=scaler.transform(x_test)
        classifier=KNeighborsClassifier(n_neighbors=len(dataset[feature].unique()))
        classifier.fit(x_train_scaled,y_train)
        y_pred=classifier.predict(x_test_scaled)
        dataset.loc[dataset[feature].isnull(),feature]=y_pred
        
```

```python
categorical_nan=[features for features in categorical_features if main_data[features].isnull().sum()>0]
categorical_nan
```

```python
dataset=main_data.copy()
x=dataset.dropna()
numerical_features=[features for features in dataset.columns if dataset[features].dtypes!='O']
x_train= x[numerical_features]
for feature in categorical_nan:
        y=dataset[feature]
        y_train=x[feature]
        y_null_index=dataset[dataset[feature].isna()].index
        x_test=dataset.loc[y_null_index][numerical_features]
        KNN (x_train,x_test,y_train,dataset,feature)
```

```python
for feature in numerical_features:
    upper_limit=dataset[feature].mean()+3*dataset[feature].std()
    lower_limit=dataset[feature].mean()-3*dataset[feature].std()
    dataset.drop(dataset[dataset[feature] <= lower_limit].index,inplace=True)
    dataset.drop(dataset[dataset[feature] >= upper_limit].index,inplace=True)
```
